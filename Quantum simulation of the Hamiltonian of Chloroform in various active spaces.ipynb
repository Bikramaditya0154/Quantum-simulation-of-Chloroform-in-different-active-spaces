```python
#Initialize molecule
import tequila as tq
geomstring="C 0.0 0.0 0.0\nH 0.0 0.0 1.073\nCl 0.0 0.0 -1.767\nCl 1.767 0.0 0.0\nCl 0.0 1.767 0.0"
molecule = tq.chemistry.Molecule(geometry = geomstring, basis_set="cc-pVDZ")
print(molecule)
```

```python
#Molecular Orbitals and Point Groups
molecule.orbitals[3]
```

```python
#Since the Hamiltonian of CHCl3 is already very large, let's Set Active Space
import tequila as tq
import numpy
active_orbitals = {"A'":[16,17,18], 'A"':[0,1,2]}
molecule = tq.chemistry.Molecule(geometry = geomstring, basis_set="cc-pVDZ", active_orbitals=active_orbitals)
H = molecule.make_hamiltonian()
eigenValues = numpy.linalg.eigvalsh(H.to_matrix())
print("Hamiltonian has {} terms".format(len(H)))
print(H)
print(eigenValues)
print("lowest energy = ", eigenValues[0])
```

```python
#Let's Set different Active Space
import tequila as tq
import numpy
active_orbitals = {"A'":[16,17,18,19,20], 'A"':[0,1,2,3]}
molecule = tq.chemistry.Molecule(geometry = geomstring, basis_set="cc-pVDZ", active_orbitals=active_orbitals)
H = molecule.make_hamiltonian()
eigenValues = numpy.linalg.eigvalsh(H.to_matrix())
print("Hamiltonian has {} terms".format(len(H)))
print(H)
print(eigenValues)
print("lowest energy = ", eigenValues[0])
```
